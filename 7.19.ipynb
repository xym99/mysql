## while 循环
- 当一个条件保持真的时候while循环重复执行语句

- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

```python
i = 0
while i <10:
    print('hahaha')
    i += 1  # i=i+1
```

# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
import random
number1 = random.randint(0,9)
number2 = random.randint(0,9)
number3 = random.randint(0,9)
number4 = random.randint(0,9)
print(number1,number2,number3,number4)
a,b,c,d = eval(input('a,b,c,d:'))
while 1:
    if (a,b,c,d) == (number1,number2,number3,number4):
        print('验证码正确')
        break
    else:
        number1 = random.randint(0,9)
        number2 = random.randint(0,9)
        number3 = random.randint(0,9)
        number4 = random.randint(0,9)
        print(number1,number2,number3,number4)
        a,b,c,d = eval(input('a,b,c,d:'))
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

```python
for i in range(10):
    if i == 5:
        break
print(i)
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(10):
    if i == 5:
        pass
    else:
        print(i)
```

# 关键字 break 和 continue
- break 跳出循环，终止循环
- continue 跳出此次循环，继续执行

```python
i = 1
while i<=10:
    print('hahaha')
    if i == 5:
        break
    else:
        i += 1
```

```python
for i in range(10):
    for j in range(10):
        print(i,j)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            print(i,j,k)
```

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            if k == 5:
                continue
            print(i,j,k)
```
